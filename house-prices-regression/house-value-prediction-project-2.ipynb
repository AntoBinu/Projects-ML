```python
for data in combine:
    data.loc[data['LotArea']>=172456,'LotArea']=0
    data.loc[data['LotArea']>=129667,'LotArea']=1
    data.loc[data['LotArea']>=86878,'LotArea']=2
    data.loc[data['LotArea']>=44089,'LotArea']=3
    data.loc[data['LotArea']>=1086,'LotArea']=4
    data.loc[data['Street']=='Pave','Street']=0
    data.loc[data['Street']=='Grvl','Street']=1
    data.loc[data['LotShape']=='IR2','LotShape']=0
    data.loc[data['LotShape']=='IR3','LotShape']=1
    data.loc[data['LotShape']=='IR1','LotShape']=2
    data.loc[data['LotShape']=='Reg','LotShape']=3
    data.loc[data['LandContour']=='HLS','LandContour']=0
    data.loc[data['LandContour']=='Low','LandContour']=1
    data.loc[data['LandContour']=='Lvl','LandContour']=2
    data.loc[data['LandContour']=='Bnk','LandContour']=3
combine=[train_df,test_df]
```

```python
x_train=np.array(train_df[['OverallQual', 'OverallCond', 'YearBuilt','BsmtQual','TotalBsmtSF','Electrical',
      'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 
       'TotRmsAbvGrd'
    
  
 
      
     
]])
x_test=np.array(test_df[['OverallQual', 'OverallCond', 'YearBuilt','BsmtQual','TotalBsmtSF','Electrical',
      'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 
       'TotRmsAbvGrd'
    
  
]].fillna(0))

y_train=np.array(train_df['SalePrice'])
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
logreg.score(x_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gdr=GradientBoostingRegressor(n_estimators=100,learning_rate=0.01,max_depth=1,random_state=0,loss='ls')
gdr.fit(x_train,y_train)
score=gdr.score(x_train,y_train)
print(score)
```

```python
y_pred=gdr.predict(x_test)
print(y_pred)
```

```python
submission=pd.DataFrame({'Id':test_df['Id'],'SalePrice':y_pred})
print(submission)
```
